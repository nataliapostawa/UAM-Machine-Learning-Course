3. Pokaż, że dla powyższych macierzy $A$ i $B$ prawdą jest, że $(AB)^T = B^TA^T$.  
4. Oblicz $\det(AB)$ (wyznacznik iloczynu $AB$).
4. Czym różni się operacja `A**-1` dla obiektów typu `array` i `matrix`? Pokaż na przykładzie. 
6. Dla macierzy $X = \left[
      \begin{array}{rrr}
        1 & 2 & 3\\
        1 & 3 & 6 \\
      \end{array}
    \right]$ oraz wektora $\vec{y} = \left[
      \begin{array}{r}
        5 \\
        6 \\
      \end{array}
    \right]$ oblicz wynikowy wektor: 
$$\vec{\theta} = (X^TX)^{-1}X^T\vec{y}.$$ Wykonaj te same obliczenia raz na obiektach typu `array` i raz na obiektach typu `matrix`. W przypadku obiektów typu `matrix` zastosuj możliwie krótki zapis. 

```python
import numpy as np

print("zad.1")
A = np.array([[1, -1, 2], [0, -3, 1]])
x = [2,1,0]

print(np.dot(A, x))

print("zad.2")
A = np.array([[0,4,-2], [-4,-3,0]])
B = np.array([[0, 1], [1, -1], [2, 3]])
print(np.dot(A, B))

print("zad.3")
C = np.dot(A, B)
print("transpozycja iloczynu:")
print(C.transpose())

print("iloczyn transpozycji:")
print(np.dot(B.transpose(), A.transpose()))
print("odpowiedź: są równe")

print("zad.4")
print(np.linalg.det(C))

print("zad.5")
print(" dla obiektu array A**-1 podnosi każdy element do potęgi (-1) a następnie rzutuje wynik na typ danych, który jest w A (int na int, float na float itp)")
print("dla obiektu matrix A**-1 oblicza macierz odwrotną do macierzy A")

print("przykład array:")
ex1 = np.array([[1,2], [3,4]])
print(ex1**-1)
print("przykład matrix:")
ex2 = np.matrix('1 2; 3 4')
print(ex2**-1)

print("zad.6")
print("array:")
X = np.array([[1,2,3], [1,3,6]])
y = np.array([5,6])
theta = np.dot(np.dot(np.linalg.inv(np.dot(X.transpose(), X)), X.transpose()), y)
print(theta)
print("matrix:")
Xm = np.matrix('1,2,3; 1,3,6')
ym = np.matrix('5; 6')
thetam = ((Xm.T*Xm)**-1)*Xm.T*ym
print(thetam)
```
